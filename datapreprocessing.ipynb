### As you see the above graph between 'n_hot_rooms' and 'price' and 'price' there are two points lied outside the group of points those are called as "OUTLIERS" which will cause the prediction of house pricing 

#### So we will see the value of points which is near to outliers ,thats why we will use the 99 percentile ,that is the right most value of points near to those outliers

```python
upper=np.percentile(hp.n_hot_rooms,[99])[0]
```

```python
upper
```

### The above value is right most value of group points

#### Now in below ,we will see the data greater than that value (upper) i.e., outliers

```python
hp[hp.n_hot_rooms>upper]
```

#### Now we equate the ouliers with 3 times of upper value we got above..we can mulitiply with 'X' times ,based on our knowledge about the datatset

```python
hp.n_hot_rooms[(hp.n_hot_rooms> 3*upper)] = 3*upper
```

```python
sb.jointplot(x='rainfall',y='price',data=hp)
```

#### As you observe that there is outliers in the graph in left side. so 1 percentile value of right side grouped data will be found below

```python
lower=np.percentile(hp.rainfall,[1])[0]
```

```python
lower
```

#### The values less than lower value will get now

```python
hp[hp.rainfall<lower]
```

#### Now we equate the ouliers with 0.5 times of lower value we got above..we can mulitiply with 'X' times ,based on our knowledge about the datatset

```python
hp.rainfall[hp.rainfall<0.5*lower]=0.5*lower
```

## Missing values

#### As you can see the above result of 'hp.info()' ,In the n_hos_beds column appears the value 498 ..i.e., data is missing data in that column.

#### Now we fill the missing data  with mean ,median and mode ..here will filling the missing data with mean value

```python
hp.n_hos_beds = hp.n_hos_beds.fillna(hp.n_hos_beds.mean())
```

Data Dictionary - House Pricing Dataset
1. price - Value of the house
2. crime_rate - Crime rate in the neighbour hood
3. resid_area - Proportion of residential area in the town
4. air_qual - Quality of air in that neighbourhood
5. room_num - Average number of rooms in houses in that locality
6. age - How old is the house construction in years
7. dist1 - Distance from employment hub 1
8. dist2 - Distance from employment hub 2
9. dist3 - Distance from employment hub 3
10. dist4 - Distance from employment hub 4
11. teachers - Number of teachers per thousand population in that town
12. poor_prop - Proportion of poor population in the town
13. airport - Is there an airport in the city(Yes/No)
14. n_hos_beds- Number of hospital beds per thousand population in town
15. n_hot_rooms- Number of hotel rooms per thousand population in town
16. waterbody - What type of natural fresh water source is there in the city(lake/river/both/none)
17. rainfall - The yearly average rainfall in centimeters
18. bus_ter - Is there a bus terminal in the city?(Yes/No)
19.parks - Proportion of land assigned as parks and green areas in town
